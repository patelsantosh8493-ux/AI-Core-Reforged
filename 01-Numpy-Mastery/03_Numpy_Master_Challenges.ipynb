```python
import numpy as np

# Task 1: Outlier Detector (औसत + 20 से बड़े नंबर्स खोजना)
data = np.array([10, 12, 11, 90, 13, 11, 95, 12])
outlier_threshold = np.mean(data) + 20
outliers = data[data > outlier_threshold]
print("Outliers:", outliers)

# Task 2: Chessboard Creator (8x8 शतरंज का पैटर्न)
chessboard = np.zeros((8, 8), dtype=int)
chessboard[::2, 1::2] = 1
chessboard[1::2, ::2] = 1
print("\nChessboard Matrix:\n", chessboard)

# Task 3: Broadcasting Magic (हर कॉलम से उसकी न्यूनतम वैल्यू घटाना)
mat = np.array([
    [10, 20, 30],
    [40, 50, 60],
    [70, 80, 90]
])
normalized_mat = mat - np.min(mat, axis=0)
print("\nNormalized Matrix:\n", normalized_mat)

# Task 4: High Earners Search (80k से ज़्यादा सैलरी वाले नाम खोजना)
names = np.array(["Amit", "Naman", "Satyam", "Rahul", "Vijay"])
salaries = np.array([45000, 95000, 35000, 120000, 50000])
hi_sal_names = names[np.where(salaries > 80000)]
print("\nHigh Earners:", hi_sal_names)

# Task 5: Vector Distance (दो 3D पॉइंट्स के बीच की यूक्लिडियन दूरी)
p1 = np.array([1, 2, 3])
p2 = np.array([4, 5, 6])
distance = np.sqrt(np.sum((p2 - p1) ** 2))
print("\nEuclidean Distance:", distance)
```
