```python
import pandas as pd

# Task 1: Method Chaining (Filter > 1000 & Sort Ascending)
df = pd.DataFrame({'Product': ['पंखा', 'कूलर', 'मिक्सर', 'टॉर्च'], 'Price': [1200, 3500, 2500, 150]})
chained_df = df[df['Price'] > 1000].sort_values(by='Price', ascending=True)
print("Task 1 - Method Chaining:\n", chained_df)

# Task 2: Dummy Variable Creator (One-Hot Encoding)
df_cat = pd.DataFrame({'Product_ID': [101, 102, 103], 'Type': ['Electric', 'Home', 'Electric']})
df_encoded = pd.get_dummies(df_cat, columns=['Type'], dtype=int)
print("\nTask 2 - One-Hot Encoded:\n", df_encoded)

# Task 3: Time-Series Resampler (3-Day Intervals)
dates = pd.date_range(start='2026-01-01', periods=7, freq='D')
sales_ts = pd.DataFrame({'Revenue': [1000, 2000, 1500, 3000, 2500, 4000, 3500]}, index=dates)
sales_resampled = sales_ts.resample('3D').sum()
print("\nTask 3 - Resampled Sales:\n", sales_resampled)

# Task 4: Duplicated Dropper (Drop Duplicates keeping Last)
df_dup = pd.DataFrame({
    'Product': ['पंखा', 'कूलर', 'पंखा', 'कूलर', 'पंखा'],
    'Price': [1200, 3500, 1200, 3500, 1300]
})
df_cleaned = df_dup.drop_duplicates(subset=['Product', 'Price'], keep='last')
print("\nTask 4 - Cleaned Duplicates:\n", df_cleaned)

# Task 5: Value Counter Percent (Market Share %)
df_bills = pd.DataFrame({'Item': ['पंखा', 'कूलर', 'पंखा', 'पंखा', 'कूलर']})
market_share = df_bills['Item'].value_counts(normalize=True) * 100
print("\nTask 5 - Market Share (%):\n", market_share)
```
