```python
import pandas as pd
import numpy as np

# Task 1: Fillna (औसत कीमत से गायब डेटा को भरना)
df_clean = pd.DataFrame({
    'Product': ['पंखा', 'कूलर', 'मिक्सर'],
    'Price': [1200, np.nan, 2500]
})
df_clean['Price'] = df_clean['Price'].fillna(df_clean['Price'].mean())
print("Task 1 - Filled NaN:\n", df_clean)

# Task 2: Groupby (कैटेगरी के अनुसार कुल रेवेन्यू का योग)
sales_df = pd.DataFrame({
    'Category': ['Electric', 'Electric', 'Home', 'Home'],
    'Product': ['पंखा', 'कूलर', 'मिक्सर', 'आयरन'],
    'Revenue': [12000, 35000, 15000, 8000]
})
category_revenue = sales_df.groupby('Category')['Revenue'].sum().reset_index()
print("\nTask 2 - Category Revenue:\n", category_revenue)

# Task 3: Lambda Transformer (1000 से ऊपर की कीमत पर 10% डिस्काउंट)
df_discount = pd.DataFrame({
    'Product': ['पंखा', 'टॉर्च', 'कूलर'],
    'Price': [1200, 150, 3500]
})
df_discount['Discounted_Price'] = df_discount['Price'].apply(lambda x: x * 0.90 if x > 1000 else x)
print("\nTask 3 - Discounted Prices:\n", df_discount)

# Task 4: Merge (दो अलग टेबल्स को आपस में जोड़ना)
df1 = pd.DataFrame({'Product': ['पंखा', 'कूलर'], 'Price': [1200, 3500]})
df2 = pd.DataFrame({'Product': ['पंखा', 'कूलर'], 'Stock': [15, 5]})
merged_df = pd.merge(df1, df2, on='Product', how='outer')
print("\nTask 4 - Merged Ledger:\n", merged_df)

# Task 5: String Clean (नाम के आगे-पीछे के फालतू स्पेस हटाना)
df_space = pd.DataFrame({'Product': ['  पंखा  ', 'कूलर  ', '  टॉर्च']})
df_space['Product'] = df_space['Product'].str.strip()
print("\nTask 5 - Cleaned Strings:\n", df_space)
```
